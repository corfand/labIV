# Introdução as ferramentas

Veremos aqui alguns exemplos da utilização das bibliotecas `numpy,pandas` e `scipy` aplicadas em um contexto simples de estatística.

Começamos importando as bibliotecas

```python
import numpy as np 
import pandas as pd
import scipy
import matplotlib.pyplot as plt
```

Utilizamos `import numpy as np`, pois no decorrer do código, como chamaremos alguns métodos do `numpy` e para não ficarmos escrevendo diversas vezes `numpy`, trocamos apenas por `np`.

Após importamos, podemos agora começar. Vamos carregar os dados lendo de um arquivo.

```python
data = pd.read_csv('data.csv')
data
```

Outros comandos auxiliam na visualização, como ` head()` e `tail()`. O primeiro comando apresenta as primeiras linhas, e o segundo comando apresenta as últimas linhas.

# Cálculo do IMC

```python
data['IMC'] = (data['weight']/(data['height'] ** 2))
data
```

```python
data.tail()
```

Através do comando `dtypes` é possível visualizar os tipos que formam o conjunto de dados.

```python
data.dtypes
```

Agora, podemos calcular as primeiras medidas estátisticas como a média e o desvio padrão.

```python
mean = np.mean(data['age'])
st_dev = np.std(data['age'])

print("média: " + str(mean))
print("desvio padrão: " + str(st_dev))
```

Encontrar a mediana, o valor máximo e mínimo também é simples.

```python
median = np.median(data['age'])
maximum = np.max(data['age'])
minimum = np.min(data['age'])

print("mediana: " + str(median))
print("valor máximo: " + str(maximum))
print("valor mínimo: " + str(minimum))
```

Uma maneira eficiente de termos um resumo das estátisticas das colunas é através do comando `describe()`.

```python
data.describe()
```

E para finalizar, podemos plotar o histograma de uma coluna, para visualizarmos a distribuição.

```python
# the histogram of the data
n, bins, patches = plt.hist(data['age'], 50, normed=1, facecolor='b', alpha=0.75)

plt.xlabel('Idade')
plt.ylabel('Porcentagem')
plt.title('Histograma de Idade')
plt.grid(True)
plt.show()
```
